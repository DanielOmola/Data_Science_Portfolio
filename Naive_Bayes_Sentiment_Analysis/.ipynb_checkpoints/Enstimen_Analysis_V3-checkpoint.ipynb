# Preprocessing

## Helper Function

```python
import regex as re
def clean(text):
    words = text.split()
    for w in words:
        if ('@' in w) or ('#' in w) : #
            words.remove(w)
        elif ('#' in w) : #
            words.remove(w)
    text= ' '.join(words)

    #stop_words = [' to ', ' gate ', ' flight ', ' fly ', ' now ', ' co ', ' will ']
    stop_words = [' to ', ' co ', ' will ']
    stop_words.extend(list(set(string.punctuation)))
    stop_words.extend([' %s '%chr(i) for i in range(97,123)])
    #stop_words.extend(stopwords.words('english'))
    #stop_words.extend([' %s '%chr(i) for i in range(97,123)])
    for w in stop_words:
        split = list(set(text.split(w)))
        text = ' '.join(split)
    return text
```

```python
def remove_emoji(text):

    emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
            u"\U0001F1F2-\U0001F1F4"  # Macau flag
            u"\U0001F1E6-\U0001F1FF"  # flags
            u"\U0001F600-\U0001F64F"
            u"\U00002702-\U000027B0"
            u"\U000024C2-\U0001F251"
            u"\U0001f926-\U0001f937"
            u"\U0001F1F2"
            u"\U0001F1F4"
            u"\U0001F620"
            u"\u200d"
            u"\u2640-\u2642"
            "]+", flags=re.UNICODE)

    #print(emoji_pattern)
    text = emoji_pattern.sub(r'', text)
    return text
```

```python
def keep_emoji(text):
    emoji_list=[]
    word_list = emoji.get_emoji_regexp().split(text)
    for w in word_list:
        w=w.strip()
        #if len(w.split())==1:
        if w in emoji.UNICODE_EMOJI['en']:
            #print(w)
            emoji_list.append(w)
    return emoji_list
```

```python

def get_word_dict(text_no_emoji,emoji_list):
    word_dict = {}
    word_list=[]

    letters = [chr(i) for i in range(97,123)]
    for w in text_no_emoji.split():
        w=w.strip()
        if (w not in letters) and (w not in stopwords.words('english')):
            if word_dict.get(w[:-1],0)!=0:
                word_dict[w[:-1]] = word_dict.get(w[:-1],0)+1
            else : word_dict[w] = word_dict.get(w,0)+1
            #word_dict[w] = word_dict.get(w,0)+1

    #print(word_dict)  

    for w in emoji_list:
        w=w.strip()
        word_dict[w] = word_dict.get(w,0)+1
    return word_dict
```

## Preprocesse the data

```python
data.text = data.text.apply(lambda t : t.lower() )
data.text = data.text.apply(lambda t : t.replace('...','') )
data.text = data.text.apply(lambda t : clean(t) )
text = ' '.join(data.text.to_list())
text_no_emoji =remove_emoji(text)
emoji_list =keep_emoji(text)
word_dict = get_word_dict(text_no_emoji,emoji_list)
```

```python
import numpy as np
# --------------- P(y) ----------------------
def prior(y=y_train):
  n = len(y) 
  priors = np.array([len(y[y == c]) / n for c in  np.unique(y)])
  #print(np.unique(y))
  return priors
#priors = prior(y_train)
#priors
```

```python
# --------------- P(Xi | y) ----------------------
def likelihood(X,y):
  alpha= 1
  n = len(y) 
  X_by_class = np.array([X[y == c] for c in np.unique(y)])
  #print(X_by_class)
  #likelihoods = np.array([len(X[y == c]) / n for c in  np.unique(y)])
  word_counts = np.array([sub_arr.sum(axis=0) + alpha for sub_arr in X_by_class])
  likelihoods = word_counts / word_counts.sum(axis=1).reshape(-1, 1)
  return likelihoods


def fit(X,y):
  priors = prior(y)
  likelihoods = likelihood(X,y)
  return priors, likelihoods

#priors, likelihoods = fit(X_train,y_train)
#likelihoods.sum(axis=1)
```

```python
def predict(X,priors, likelihoods):
    lkh = np.multiply(X.T, likelihoods)
    lkh_ = [lkh[i][lkh[i]>0]*1 for i in range(lkh.shape[0])]
    #print(lkh_)
    Pxy =np.prod(lkh_,axis=1)
    Pxy
    #np.product(np.multiply(lkh,axis = 1)
    predict = np.argmax(np.prod([priors,Pxy],axis=0))
    return predict
```

```python
def evaluate(X,y,priors,likelihoods):
    good_prediction = 0
    total_prediction = 0
    for ind in range(0,X.shape[0]-1):
        total_prediction+=1
        x=np.array(X.iloc[ind].to_list())
        x=np.repeat(x, 3, axis=0)
        x=x.reshape(-1,3)

        target_string = ['negative', 'neutral', 'positive']
        #print(target_string[predict(x,priors, likelihoods)],y.iloc[ind])
        if predict(x,priors, likelihoods) == target_string.index(y.iloc[ind]):
            good_prediction+=1
    return good_prediction/total_prediction * 100       
```

```python
def reshape(x):
    x=np.array(x.to_list())
    x=np.repeat(x, 3, axis=0)
    x=x.reshape(-1,3)
    return x
```

```python
def get_predictions(X,y):
    predictions = X.apply(lambda x :predict(reshape(x),priors, likelihoods) ,axis=1)
    predictions=predictions.apply(lambda x : ['negative', 'neutral', 'positive'][x])
    predictions = pd.concat([predictions,y],axis=1)
    predictions.columns = ['prediction','target']
    return predictions
```

```python
def performance(y_target,y_predict):
    print("Accuracy:",metrics.accuracy_score(y_target, y_predict)*100)
    print("Recall:",recall_score(y_target, y_predict, average='macro')*100)
    print("Precision:",precision_score(y_target, y_predict, average='macro')*100)
```

# Feature Importance

```python
priors, likelihoods = fit(X_train,y_train)
accuracy_i = evaluate(X_train,y_train,priors, likelihoods)

number_of_columns = 50
accuracy_list = []
for c in X_train.columns[:number_of_columns]:
    X_temp = X_train.copy()
    X_temp = X_temp.drop(columns=[c])
    priors, likelihoods = fit(X_temp,y_train)
    accuracy_t = evaluate(X_temp,y_train,priors, likelihoods)
    accuracy_list.append(accuracy_t)
    

accuracy_DF = pd.DataFrame({'feature':X_train.columns[:number_of_columns],
                           'accuracy' : accuracy_list })
accuracy_DF['initial_accuracy'] = accuracy_i
accuracy_DF['accuracy_gain'] = accuracy_DF['accuracy']-accuracy_i
accuracy_DF
```

```python
pat1 = r'@[A-Za-z0-9]+' # this is to remove any text with @....
pat2 = r'https?://[A-Za-z0-9./]+'  # this is to remove the urls
combined_pat = r'|'.join((pat1, pat2)) 
pat3 = r'[^a-zA-Z]' # to remove every other character except a-z & A-Z
combined_pat2 = r'|'.join((combined_pat,pat3)) # we combine pat1, pat2 and pat3 to pass it in the cleaning steps
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
cleaned_tweets = []

for i in range(0, len(data['text'])) :
    tweets = re.sub(combined_pat2,' ',data['text'][i])

    tweets = tweets.lower()
    tweets = tweets.split()

    tweets = [ps.stem(word) for word in tweets if not word in set(stopwords.words('english'))]
    tweets = ' '.join(tweets)
    cleaned_tweets.append(tweets)
```
