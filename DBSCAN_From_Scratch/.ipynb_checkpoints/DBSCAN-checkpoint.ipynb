# 0 - Modules

```python
# ##################################
#
#  much of the cod is stored in 
#  my package for more readability
#
# ################################## 
from mypackage import clustering
from mypackage import ploter as plt
from mypackage import datagenerator as dg

import numpy as np
from sklearn import  datasets
n_samples = 1500
```

# 1 - DBSCAN on spherical dataset

## 1.1 - Load Data

```python
data = dg.GenerateData()
plt.plot_scatter(data,title='Spherical Dataset')
```

## 1.2 - Train the model and plot the result

```python
dbscan = clustering.DBSCAN(eps = 2,Min_neighbor = 30)
Clusters_List,cluster_labels = dbscan.fit(data)
plt.plot_scatter(data,cluster_labels.tolist(),title='DBSCAN on Spherical Dataset')
```

# 2 - DBSCAN on non spherical shape clusters

## 2.1 - Noisy Circles

```python
dbscan = clustering.DBSCAN(eps = .2,Min_neighbor = 55)
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,noise=.05)
data = noisy_circles[0]
Clusters_List,cluster_labels = dbscan.fit(data)
plt.plot_scatter(data,cluster_labels.tolist(),title='DBSCAN on Noisy Circles')
```

## 2.2 - Blobs

```python
dbscan = clustering.DBSCAN(eps = 1,Min_neighbor = 50)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
data = blobs[0]
Clusters_List,cluster_labels = dbscan.fit(data)
plt.plot_scatter(data,cluster_labels.tolist(),title='DBSCAN on Blobs')
```

## 2.3 - Data without any structure

```python
dbscan = clustering.DBSCAN(eps = 1,Min_neighbor = 50)
no_structure = np.random.rand(n_samples, 2), None
data = no_structure[0]
Clusters_List,cluster_labels = dbscan.fit(data)
plt.plot_scatter(data,cluster_labels.tolist(),title='DBSCAN on data without any structure')
```

## 2.4 - Noisy Moons

```python
dbscan = clustering.DBSCAN(eps = .2,Min_neighbor = 50)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
data = noisy_moons[0]
Clusters_List,cluster_labels = dbscan.fit(data)
plt.plot_scatter(data,cluster_labels.tolist(),title='DBSCAN on Noisy Moons')
```
